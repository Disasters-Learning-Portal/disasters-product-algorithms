# Satellogic Processing Workflow #

This notebook demonstrates the complete workflow for downloading and processing Satellogic imagery using the `disasters-product-algorithms` package.

## Workflow Steps
1. **Configure Environment Variables** - Set processing parameters
2. **Process Satellogic Data** - Generate products with COG conversion
3. **View Results** - Examine the generated outputs

## Features Demonstrated
- Cloud Optimized GeoTIFF (COG) conversion
- Multiple product generation (true color, color IR, NDVI, NDWI)
- Cloud masking (L2A only)

# Environment Setup #

Configure all processing parameters as environment variables for easy modification.

```python
import os
import subprocess

DATE = "2024-06-01 12:00:00"
LEVEL = "L1D_SR_MS"
PRODUCTS = ["colorir"]   # truecolor, ndvi, ndwi, colorir
OUTPUT_DIR = "./s3_temp"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# COG options
COMPRESSION = "ZSTD"
COMPRESSION_LEVEL = 22
NODATA = 0

print("Configuration:")
print(f"  Date: {DATE}")
print(f"  Level: {LEVEL}")
print(f"  Product: {PRODUCTS}")
print(f"  Output Directory: {OUTPUT_DIR}")
```

# Process Satellogic Data #

Process the downloaded imagery to generate various products with COG conversion and event naming.

**Note:** The processing script has been configured to display progress in real-time within JupyterHub. You'll see:
- Detailed product generation steps
- COG conversion progress
- Error messages if any products fail
- Final processing summary with success/failure counts
- Log file location for detailed error tracking

```python
process_cmd = [
    "process_satellogic",
    "-h"
]

help_flags = subprocess.run(process_cmd, cwd=os.getcwd())
print(help_flags)
```

```python
for PRODUCT in PRODUCTS:
    process_cmd = [
        "process_satellogic",
        "--product", PRODUCT,
        "--date", DATE,
        "--level", LEVEL,
        "--output", OUTPUT_DIR,
        "-compression", COMPRESSION,
        "-compression_level", str(COMPRESSION_LEVEL)
    ]
    
    # Add optional parameters
    if NODATA is not None:
        process_cmd.extend(["-nodata", str(NODATA)])
    
    print(f"Processing Satellogic data: {PRODUCT}...")
    print(f"Command: {' '.join(process_cmd)}")
    print()
    
    process = subprocess.Popen(
        process_cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        universal_newlines=True,
        bufsize=1
    )
    
    for line in process.stdout:
        print(line, end='')
    
    return_code = process.wait()
    
    if return_code == 0:
        print("\n✓ Processing completed successfully!")
    else:
        print(f"\n✗ Processing failed with return code {return_code}")
```

# View Results #

Examine the generated output files and directory structure.

```python
import glob
import os
import rasterio
import matplotlib.pyplot as plt
import numpy as np

output_dir = os.path.abspath(OUTPUT_DIR)

if output_dir:
    print(f"✓ Found output directory: {output_dir}\n")
    
    # Find all tifs
    tif_files = sorted([
        f for f in glob.glob(os.path.join(output_dir, "**/*.tif"), recursive=True)
        if not f.endswith('.tmp.tif')
    ])
    
    if tif_files:
        # Group by product type
        product_types = {}
        for tif_file in tif_files:
            product_type = os.path.basename(os.path.dirname(tif_file))
            product_types.setdefault(product_type, []).append(tif_file)

        print(f"Displaying {len(product_types)} product types...\n")

        # Plotting
        for tif_file in tif_files:
            print(f"\nDisplaying: {os.path.basename(tif_file)}")
        
            with rasterio.open(tif_file) as src:
        
                plt.figure(figsize=(8, 8))
        
                # RGB Products
                if src.count >= 3:
                    arr = src.read([1, 2, 3]).astype(float)
                    print(arr.dtype)
                    print(arr.min(), arr.max())
                    arr = np.transpose(arr, (1, 2, 0))

                    # Compute min/max per band
                    band_min = np.min(arr, axis=(0, 1))
                    band_max = np.max(arr, axis=(0, 1))
                    
                    # Normalize each band to 0-1
                    arr = (arr - band_min) / (band_max - band_min + 1e-6)
                    
                    # Ensure valid range
                    arr = np.clip(arr, 0, 1)
                    
                    plt.imshow(arr)

                    pixels = arr.reshape(-1, 3)

                    plt.figure(figsize=(8, 4))
                    plt.hist(pixels[:, 0], bins=256, range=(0.01, 1), color='red', alpha=0.5, label='R')
                    plt.hist(pixels[:, 1], bins=256, range=(0.01, 1), color='green', alpha=0.5, label='G')
                    plt.hist(pixels[:, 2], bins=256, range=(0.01, 1), color='blue', alpha=0.5, label='B')
                    
                    plt.title("RGB Channel Histograms")
                    plt.xlabel("Pixel value (0–255)")
                    plt.ylabel("Count")
                    plt.legend()
                    plt.show()
        
                # Single Band Products
                else:
                    arr = src.read(1).astype(float)
        
                    filename = os.path.basename(tif_file).lower()
        
                    if "ndvi" in filename:
                        cmap = "RdYlGn"
        
                        # Mask invalid values
                        arr = np.where((arr < -1) | (arr > 1), np.nan, arr)
        
                        im = plt.imshow(arr, cmap=cmap, vmin=-0.5, vmax=0.5) # Add vmin and vmax for colorbar how you'd like to enhance contrast of image
        
                    elif "ndwi" in filename:
                        cmap = "Blues"

                        arr = np.where((arr < -1) | (arr > 1), np.nan, arr)
                        
                        im = plt.imshow(arr, cmap=cmap)
        
                    else:
                        im = plt.imshow(arr, cmap="viridis")
        
                    plt.colorbar(im, fraction=0.046, pad=0.04)
        
                plt.title(os.path.basename(tif_file))
                plt.axis('off')
                plt.tight_layout()
                plt.show()
        
                print(f"✓ Plotted {os.path.basename(tif_file)}")

            plt.tight_layout()
            plt.show()

        # File details
        print("\nFile Details:")
        print("=" * 50)
        for p_type, files in sorted(product_types.items()):
            print(f"\n{p_type}:")
            for f in files:
                size = os.path.getsize(f) / (1024 * 1024)
                print(f"  - {os.path.basename(f)} ({size:.1f} MB)")
    else:
        print("No finished GeoTIFF files found. Check if the processing script is still running.")
else:
    print(f"Output directory not found.")
```
